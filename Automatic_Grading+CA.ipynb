Bag of Words + Correspondance Analysis

```python
documents = [clean_text1,clean_text2,clean_text3,clean_text4]
```

```python
word2count= {}
for data in documents:
    words = nltk.word_tokenize(data)
    for word in words:
        if word not in word2count.keys():
            word2count[word]= 1
        else:
            word2count[word] +=1
            
```

```python
word2count
```

```python
len(word2count)
```

```python
import heapq
freq_words = heapq.nlargest(42,word2count,key=word2count.get)
```

```python
x=[]
for data in documents:
    vector= []
    for word in freq_words:
        if word in nltk.word_tokenize(data):
            vector.append(1)
        else:
            vector.append(0)
    x.append(vector)
    
    
```
