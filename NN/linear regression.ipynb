```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = [2,4,5,7,8,9,10]
y = [3,6,6,7,9,10,11]
```

```python
plt.scatter(x,y)
```

```python
alpha = 0.001
iterations = 1000
x0 = np.ones(len(x))
X = np.array([x0,x]).T
```

```python
def cost(X,y, theta=[[0],[0]]):
    return (np.sum(np.square((X.dot(theta) - y)))/(2 * len(y)))
```

```python
theta = np.array([0,0]) 
print(cost(X,y,theta))
```

```python
def gradient(X,y,theta,alpha,iterations):
    cost_hist = np.zeros(iterations)
    for i in range(iterations):
        theta =  theta - alpha*((X.T.dot(X.dot(theta)-y))/len(y))
        cost_hist[i] = cost(X,y,theta)
    return theta,cost_hist
```

```python
f_theta,history = gradient(X,y,theta,alpha,iterations)
print(f_theta)
```

```python
plt.plot(history)
```

```python
y_final = [(i*f_theta[1] + theta[0]) for i in x]
```
